```python
import openpyxl
from openpyxl.styles import Font, Border, Side
import random
from openpyxl.styles import Alignment

# 既存のエクセルファイルをオープン
wb = openpyxl.load_workbook('../../../Downloads/train_data2.xlsx')
sheet = wb.active
random.seed(42) 
# セルのサイズを調整して正方形にする
cell_size = 20  # セルの一辺のサイズ（ピクセル数）
# for i in range(1, 101):
#     sheet.column_dimensions[openpyxl.utils.get_column_letter(i)].width = cell_size
#     sheet.row_dimensions[i].height = cell_size

# フォントのリストを定義
font_list = [
    "Arial", "Times New Roman", "Calibri", "Verdana", "Tahoma", "Courier New",
    "Segoe UI", "Comic Sans MS", "Impact", "Georgia", "Palatino Linotype",
    "Lucida Console", "Trebuchet MS", "Bookman Old Style", "MS Sans Serif",
    "Century Gothic", "Franklin Gothic Medium", "Ebrima", "Cambria", "Constantia",
    "メイリオ", "游ゴシック", "ＭＳ Ｐゴシック", "ＭＳ ゴシック", "ＭＳ Ｐ明朝", "ＭＳ 明朝",
    "ヒラギノ角ゴシック", "ヒラギノ角明朝", "Osaka", "IPA明朝", "IPAゴシック", "VL ゴシック",
    "Noto Sans CJK JP", "Noto Serif CJK JP", "Hiragino Kaku Gothic Pro", "Hiragino Maru Gothic Pro",
    "HG創英角ﾎﾟｯﾌﾟ体", "HGｺﾞｼｯｸM", "ＤＦＰ勘亭流", "ＤＦＰ中太明朝", "ＤＦＰ平成明朝", "ＤＦＰ特太明朝",
]

# セルの境界線を削除
border = Border(left=Side(border_style=None), 
                right=Side(border_style=None), 
                top=Side(border_style=None), 
                bottom=Side(border_style=None))

for row in sheet.iter_rows(min_row=1, max_row=100, min_col=1, max_col=100):
    for cell in row:
        cell.border = border

# 正解データとランダムな数字を配置するセル範囲
rows = 100
cols = 100
data_range = sheet.iter_rows(min_row=1, max_row=rows, min_col=1, max_col=cols)

# 100x100のセルにランダムな数字を書き込み、フォントを設定
for row_index, row in enumerate(data_range, start=1):
    for col_index, cell in enumerate(row, start=1):
        number = random.randint(1, 9)
        cell.value = number
        selected_font = random.choice(font_list)
        cell.font = Font(name=selected_font,size=18)
for row in sheet.iter_rows():
    for cell in row:
        horizontal_alignment = random.choice(["center", "left", "right"])
        vertical_alignment = random.choice(["center", "top", "bottom"])
        # horizontal_alignment="center"
        # vertical_alignment="center"
        alignment = Alignment(horizontal=horizontal_alignment, vertical=vertical_alignment)
        cell.alignment = alignment
# エクセルファイルを保存
wb.save('../../../Downloads/train_data2.xlsx')
```
